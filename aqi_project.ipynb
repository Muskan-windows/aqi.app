```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['datetime'], df['temp_c'])
plt.title('Temperature over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['datetime'], df['aqi_index'])
plt.title('AQI Trend')
plt.show()
```

```python
df[['temp_c', 'humidity', 'pm2_5', 'pm10', 'aqi_index']].corr()
```

```python
df.set_index('datetime', inplace=True)

df['aqi_index'].resample('D').mean()
```

```python

df = df.drop(columns=['date_ist', 'time_ist'])
```

```python
#time feature
df['hour'] = df.index.hour
df['day'] = df.index.day_name()
df['month'] = df.index.month
```

```python
#days aqi is high bcoz pollution
#Aqi by hour
df.groupby('hour')['aqi_index'].mean().plot()
plt.title("AQI by Hour")
plt.show()
```

```python
#weekdays are more polluted
#aqi by day
df.groupby('day')['aqi_index'].mean().plot()
```

```python
#aqi vs pm2.5
plt.scatter(df['pm2_5'], df['aqi_index'])
plt.xlabel("PM2.5")
plt.ylabel("AQI")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: select relevant columns
df_clean = df[['pm2_5', 'pm10', 'no2', 'co', 'aqi_index']]

# Step 2: convert to numeric (safety)
df_clean = df_clean.apply(pd.to_numeric, errors='coerce')

# Step 3: remove missing values
df_clean = df_clean.dropna()

# Step 4: split features and target
X = df_clean[['pm2_5', 'pm10', 'no2', 'co']]
y = df_clean['aqi_index']

# Step 5: train model
model = LinearRegression()
model.fit(X, y)

print("R2 Score:", model.score(X, y))
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_clean['pm2_5'], df_clean['aqi_index'])
plt.xlabel("PM2.5")
plt.ylabel("AQI Index")
plt.title("PM2.5 vs AQI")
plt.show()
```
